```python
# quantization
import numpy as np
def int16_to_float32(x):
    return (x / 32767.0).astype(np.float32)


def float32_to_int16(x):
    x = np.clip(x, a_min=-1., a_max=1.)
    return (x * 32767.).astype(np.int16)
```

```python
ma = 'HTSAT-base'
clap_model = laion_clap.CLAP_Module(enable_fusion=False, amodel=ma)
clap_model.load_ckpt('/trunk/code/kechen/CLAP-pip-test/music_audioset_epoch_15_esc_90.14.pt')
```

```python
audio_file = [
    'test_clap_long.wav',
    # '/trunk/code/kechen/CLAP-pip-test/test_clap_long.wav',
    # '/trunk/code/kechen/CLAP-pip-test/test_clap_short.wav',
    # '/trunk/code/kechen/CLAP-pip-test/test_clap_short.wav',
    # '/trunk/code/kechen/CLAP-pip-test/test_clap_short.wav',
    'test_clap_short.wav'
]
audio_embedding = clap_model.get_audio_embedding_from_filelist(x = audio_file, use_tensor=False)
print(audio_embedding[:, -20:])
print(audio_embedding.shape)
```
